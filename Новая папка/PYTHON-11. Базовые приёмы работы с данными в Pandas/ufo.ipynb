```python
import pandas as pd

ufo_data = pd.read_csv('https://raw.githubusercontent.com/justmarkham/pandas-videos/master/data/ufo.csv', sep=',')
ufo_df = ufo_data.copy()
display(ufo_df)
```

В каком году отмечается наибольшее количество случаев наблюдения НЛО в США?

```python
ufo_df['Time'] = pd.to_datetime(ufo_df['Time'])
display(ufo_df)
date_ufo = ufo_df['Time'].dt.year.value_counts()
print(date_ufo)
```

Найдите средний интервал времени (в днях) между двумя последовательными случаями наблюдения НЛО в штате Невада (NV).

```python
ufo_df[ufo_df['State'] == 'NV']['Time'].dt.date.diff().mean()
```
