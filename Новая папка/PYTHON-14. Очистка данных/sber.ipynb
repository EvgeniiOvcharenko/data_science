```python
fig = plt.figure(figsize = (10,7))
box = sns.boxplot(data = sber_data, y = 'ecology', x = 'price_doc', orient = 'h')
box.set_title(' коробчатую диаграмму цен на квартиры (price_doc) в зависимости от уровня экологической обстановки в районе (ecology)')
box.set_xlabel('Цены')
box.set_ylabel('Экологическая обстановка')
box.grid()
```
